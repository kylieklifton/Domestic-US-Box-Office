In this section we are going to download Domestic US box office data from IMDB Charts and Box Office Mojo which is a product of IMDB Pro. They will be saved as seperate CSV files.

The first URL is https://www.imdb.com/chart/boxoffice/

The second URL is https://www.boxofficemojo.com/date/?ref_=bo_nb_wey_secondarytab

```python
# Normally we'd do this...
import requests
from bs4 import BeautifulSoup

#... but this we'll just do this!
import pandas as pd
```

```python
tables = pd.read_html("https://www.boxofficemojo.com/date/?ref_=bo_nb_wey_secondarytab")
len(tables)
```

```python
df = tables[0]
df.head()
```

```python
df = df.drop(columns=['Day #', '%± LW'])
```

```python
df = df.rename(columns={
    'Releases': 'Number of Films in Release'
})
```

```python
df = df.rename(columns={
    '%± YD': 'Percent Change vs Prior Day'
})
df.head()
```

```python
df.to_csv("box-office.csv", index=False)
```
